#### 读取LHA neurons

```python
iondata=IONData.IONData()
r=open('../resource/LHA_neurons.pkl','rb')
LHAneuron=pickle.load(r)
LHAneuron
```

#### 获取neuron projectregion 属性

```python
LHAlength=[]
LHAlength=pd.DataFrame()
for neuron in LHAneuron:
    property = iondata.getNeuronPropertyByID(neuron[:-8],neuron[-7:])
    length=pd.DataFrame([property['projectregion']])
    LHAlength=LHAlength.append(length)
```

```python
LHAlength.index=LHAneuron
LHAlength
```

#### neuron投射脑区筛选：滤除length小于500的

```python
LHAlength1=LHAlength.copy(deep=True)
##滤除axon length小于500的
LHAlength1=LHAlength1.fillna(0)
LHAlength1[LHAlength1<500]=0
```

```python
LHAlength1.loc['sum'] =LHAlength1.apply(lambda x:x.sum())#按列
##去除总投射为0的脑区
for i in LHAlength1.columns:
    if LHAlength1[i]['sum']==0:
        LHAlength1.pop(i)
```

```python
LHAlength1.drop(index ='sum',inplace = True)
LHAlength1
```

#### 画图

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
import scipy.stats
from scipy.cluster.hierarchy import dendrogram, linkage,fcluster
%matplotlib inline
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
# LHAlength2=LHAlength1
LHAlength2=(np.log2(LHAlength1/1000.0+1))
LHAlength2
```

```python
spearman=LHAlength2.T.corr('spearman')
spearman=spearman.fillna(0)
spearman
```

```python

Z = linkage(spearman, method='ward')
f = fcluster(Z,t=19,criterion='distance')
f
```

```python

df=pd.DataFrame({'labels':f,'varieties':spearman.index})
spearman['s']=df['labels'].tolist()
row_c = dict(zip(spearman['s'].unique(), ['yellowgreen','red','pink','skyblue','yellow','mediumpurple','teal','orange','green','blue','c','indigo']))
row_c
```

```python
df=pd.DataFrame({'labels':f,'varieties':spearman.index})
spearman['s']=df['labels'].tolist()
row_c = dict(zip(spearman['s'].unique(), ['yellowgreen','red','pink','skyblue','yellow','mediumpurple','teal','orange','green','blue','c','indigo']))
row_colors = spearman['s'].map(row_c) 
row_colors
```

```python

spearman.pop('s')
sns.set(font_scale=1)
fig = plt.figure(figsize=(40,40))
cg1=sns.clustermap(spearman,fmt="d",method='ward',cmap='Spectral',center =0,row_cluster=True,
               col_cluster=True,xticklabels=False,yticklabels=False,col_colors=row_colors)          
cg1.ax_row_dendrogram.set_visible(False)#删除左侧树状图
# plt.savefig('../resource/lha_heat_cluster.pdf',format='pdf',dpi=800,transparent=True, bbox_inches='tight', pad_inches=0.0)   
plt.show()
```
