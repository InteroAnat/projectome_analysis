```python
import csv
#open的第一个参数为文件名，第二个参数为打开的类型"r"为读,'w'为写，'a'为追加等等
csvFile = open("../resource/10_slice/slice.csv","r")
reader = csv.reader(csvFile)
#将reader转化为list,方便进行遍历
slice2allen =list(reader)
```

```python
position=[]
file=[]

for i in slice2allen:
    position.append(int(i[1]))
    file.append('../resource/10_slice/T'+i[0]+'.png')
```

```python
file
```

```python
from  RenderGL import *
from skimage import draw, io

render = RenderGL(renderModel=0)
render.setBackgroundColor([1,1,1])
for i in range(len(position[:])):
    ga = GeometryAdapter.GeometryAdapter()
    ga.geometry.addPoint([position[i]*10,0,11400],color=[1.0,1.0,1.0],texcoord = [0.11,0.07])
    ga.geometry.addPoint([position[i]*10,0,5700],color=[1.0,1.0,.0],texcoord = [0.6,0.07])
    ga.geometry.addPoint([position[i]*10,8000,5700],color=[1.0,1.0,.0],texcoord = [0.6,0.71])
    ga.geometry.addPoint([position[i]*10,8000,11400],color=[1.0,1.0,.0],texcoord = [0.11,0.71])

    ga.geometry.addIndex(0)
    ga.geometry.addIndex(1)
    ga.geometry.addIndex(2)
    ga.geometry.addIndex(2)
    ga.geometry.addIndex(3)
    ga.geometry.addIndex(0)

    ga.geometry.drawModel='normal'
    img=io.imread(file[i])
    ga.geometry.addTexture('diff',img)
    # render.setPointSize(2)
    render.addGeometry(ga.geometry)
# render.setClip(True)
# render.savepng('../resource/test3.png')
render.run()
```
